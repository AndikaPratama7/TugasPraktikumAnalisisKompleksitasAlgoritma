**Carilah jumlah semua elemen dalam sebuah daftar/array.**

```python
# n->number of elements in array
def listSum(arr, n):
    sum = 0
    for i in range(n):
        sum = sum + arr[i]
    return sum

arr = [5, 6, 1, 2]
n = len(arr)
print(listSum(arr, n))
```

**Carilah jumlah semua elemen dari sebuah matriks.**

```python
n = 3
m = 3
arr = [[3, 2, 7], [2, 6, 8], [5, 1, 9]]
sum = 0

# Melakukan iterasi pada semua array 1-D dalam array 2-D.
for i in range(n):

    # Mencetak semua elemen dalam array 1-D ke-i
    for j in range(m):

        # Mencetak elemen ke-j dari baris ke-i
        sum += arr[i][j]

print(sum)
```

Program ini melakukan iterasi melalui semua elemen dalam array 2D menggunakan dua loop bersarang. Loop luar melakukan iterasi sebanyak n kali dan loop dalam melakukan iterasi sebanyak m kali untuk setiap iterasi loop luar. Oleh karena itu, kompleksitas waktu program ini adalah O(n*m).

```python
# CONTOH O(Log N)

def binary_search(sorted_list, target):
    low = 0
    high = len(sorted_list) - 1

    while low <= high:
        mid = (low + high) // 2  # Temukan indeks tengah
        guess = sorted_list[mid]

        if guess == target:
            return mid  # Target ditemukan, kembalikan indeks.
        if guess > target:
            high = mid - 1  # Targetnya berada di bagian kiri.
        else:
            low = mid + 1   # Targetnya berada di bagian kanan.

    return -1  # Target tidak ada dalam daftar

# --- Contoh Input dan Output ---
my_list = [1, 3, 5, 7, 9, 11, 13, 15]
target_val = 7

result = binary_search(my_list, target_val)

if result != -1:
    print(f"Input: {my_list}, Target: {target_val}")
    print(f"Output: Element ditemukan dalam index {result}")
else:
    print("Output: Element tidak ditemukan")
```

Langkah Awal: Anda mulai dengan elemen-elemen.

Langkah 1: Anda memeriksa bagian tengah dan membuang setengahnya. Sekarang Anda memiliki n/2 elemen yang tersisa.

Langkah 2: Anda membuang setengah lainnya. Sekarang Anda memiliki
n/4 elemen yang tersisa.

Jika n=16, jumlah maksimum langkah adalah 4. (Log 16=4)
