**Pseudecode RED BLACK TREE**

FUNGSI Insert(root, data_baru):
    node_baru = Buat Node(data_baru, warna=MERAH, left=NULL, right=NULL)
    
    JIKA root adalah NULL:
        node_baru.warna = HITAM
        KEMBALIKAN node_baru
        
    parent = NULL
    current = root
    

    SELAMA current TIDAK NULL:
        parent = current
        JIKA data_baru < current.data:
            current = current.left
        SEBALIKNYA:
            current = current.right
            
    node_baru.parent = parent
    JIKA data_baru < parent.data:
        parent.left = node_baru
    SEBALIKNYA:
        parent.right = node_baru
        

    KEMBALIKAN Insert_Fixup(root, node_baru)

```python
# ==========================================
# IMPLEMENTASI RED-BLACK TREE (TANPA CLASS)
# ==========================================

def create_node(data):
    """Membuat node baru sebagai dictionary"""
    return {
        'data': data,
        'color': 'RED', # Node baru selalu merah pada awalnya
        'left': None,
        'right': None,
        'parent': None
    }

def get_color(node):
    """Mendapatkan warna node. Null/None dianggap HITAM."""
    if node is None:
        return 'BLACK'
    return node['color']

def left_rotate(root, x):
    """Rotasi ke kiri pada node x"""
    y = x['right']
    x['right'] = y['left']
    if y['left'] is not None:
        y['left']['parent'] = x

    y['parent'] = x['parent']
    if x['parent'] is None:
        root = y
    elif x == x['parent']['left']:
        x['parent']['left'] = y
    else:
        x['parent']['right'] = y

    y['left'] = x
    x['parent'] = y
    return root

def right_rotate(root, y):
    """Rotasi ke kanan pada node y"""
    x = y['left']
    y['left'] = x['right']
    if x['right'] is not None:
        x['right']['parent'] = y

    x['parent'] = y['parent']
    if y['parent'] is None:
        root = x
    elif y == y['parent']['right']:
        y['parent']['right'] = x
    else:
        y['parent']['left'] = x

    x['right'] = y
    y['parent'] = x
    return root

def insert_fixup(root, z):
    """Memperbaiki tree jika ada pelanggaran rule setelah insert"""
    while z['parent'] is not None and get_color(z['parent']) == 'RED':
        if z['parent'] == z['parent']['parent']['left']:
            y = z['parent']['parent']['right']

            if get_color(y) == 'RED':
                z['parent']['color'] = 'BLACK'
                y['color'] = 'BLACK'
                z['parent']['parent']['color'] = 'RED'
                z = z['parent']['parent']
            else:
                if z == z['parent']['right']:
                    z = z['parent']
                    root = left_rotate(root, z)
                z['parent']['color'] = 'BLACK'
                z['parent']['parent']['color'] = 'RED'
                root = right_rotate(root, z['parent']['parent'])
        else:
            y = z['parent']['parent']['left']
            if get_color(y) == 'RED':
                z['parent']['color'] = 'BLACK'
                y['color'] = 'BLACK'
                z['parent']['parent']['color'] = 'RED'
                z = z['parent']['parent']
            else:
                if z == z['parent']['left']:
                    z = z['parent']
                    root = right_rotate(root, z)
                z['parent']['color'] = 'BLACK'
                z['parent']['parent']['color'] = 'RED'
                root = left_rotate(root, z['parent']['parent'])

    root['color'] = 'BLACK'
    return root

def insert(root, data):
    """Fungsi utama untuk memasukkan data ke dalam tree"""
    z = create_node(data)
    y = None
    x = root


    while x is not None:
        y = x
        if z['data'] < x['data']:
            x = x['left']
        else:
            x = x['right']

    z['parent'] = y
    if y is None:
        root = z
    elif z['data'] < y['data']:
        y['left'] = z
    else:
        y['right'] = z


    if z['parent'] is None:
        z['color'] = 'BLACK'
        return root

    if z['parent']['parent'] is None:
        return root

    return insert_fixup(root, z)
```

**Worst Case** Terjadi ketika struktur pohon mencapai tinggi maksimumnya. Meskipun Red-Black Tree adalah Self-Balancing, tinggi pohon ($h$) tetap dibatasi secara matematis agar tidak melebihi $2 \log_2(n+1)$. Operasi pencarian harus menelusuri jalur dari akar hingga daun terdalam.Kompleksitas: $O(\log n)$

**Average Case** Terjadi ketika elemen-elemen didistribusikan secara acak dalam tabel/pohon. Secara statistik, tinggi pohon akan mendekati $\log_2 n$. Karena operasi pencarian berbanding lurus dengan tinggi pohon, maka waktu yang dibutuhkan tetap logaritmik.Kompleksitas: $O(\log n)$
